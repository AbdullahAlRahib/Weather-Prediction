```python
#datset import
import pandas as pd
dataset = pd.read_csv('weather.numeric.csv')
```

```python
dataset
```

```python
dataset.shape
```

```python
dataset.columns
```

```python
feature_variable = ['outlook','temperature','humidity','windy']
class_variable = ['play']
```

```python
feature_variable
```

```python
x = dataset[feature_variable]
y = dataset[class_variable]
```

```python
y
```

```python
x
```

```python
#Encoding categorical data
import numpy as np
from sklearn.compose import ColumnTransformer
from sklearn.preprocessing import OneHotEncoder
ct = ColumnTransformer(transformers=[('encoder', OneHotEncoder(), ['outlook'])], remainder='passthrough')
x = np.array(ct.fit_transform(x))
```

```python
print(x)
```

```python
#Splitting train and test set
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x,y, test_size = 0.60, random_state = 0)
```

```python
x_test
```

```python
#Decision tree Model build 

from sklearn.tree import DecisionTreeClassifier
play_decision = DecisionTreeClassifier(max_leaf_nodes = 30, random_state = 0)
play_decision.fit(x_train,y_train)
```

```python
#Apply test set on built model 
prediction = play_decision.predict(x_test)
```

```python
prediction
```

```python
#To see the accuracy

from sklearn.metrics import accuracy_score
from sklearn.metrics import confusion_matrix

accuracy = accuracy_score(y_test,prediction)
accuracy*100
```

**When we are using  split 60.0% train, remainder test data we get maximum accuracy when we are using Decision tree Algorithm**

```python
#confusion matrix
c_matrix = confusion_matrix(y_test,prediction)
c_matrix
```
