```python
#import datset
import pandas as pd
dataset = pd.read_csv('weather.numeric.csv')
```

```python
dataset
```

```python
dataset.shape
```

```python
dataset.columns
```

```python
feature_variable = ['outlook','temperature','humidity','windy']
class_variable = ['play']
```

```python
feature_variable
```

```python
x = dataset[feature_variable]
y = dataset[class_variable]
```

```python
y
```

```python
x
```

```python
#Encoding categorical data
import numpy as np
from sklearn.compose import ColumnTransformer
from sklearn.preprocessing import OneHotEncoder
ct = ColumnTransformer(transformers=[('encoder', OneHotEncoder(), ['outlook'])], remainder='passthrough')
x = np.array(ct.fit_transform(x))
```

```python
print(x)
```

```python
#Splitting train and test set
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x,y, test_size = 0.50, random_state = 0)
```

```python
x_test
```

```python
from sklearn.preprocessing import StandardScaler
```

```python
scaler = StandardScaler()
scaler.fit(x_train)
```

```python
x_test = scaler.transform(x_test)
```

```python
#Knn classifier Model Bulding
from sklearn.neighbors import KNeighborsClassifier
Classifier = KNeighborsClassifier(n_neighbors=5)
Classifier.fit(x_train,y_train)
```

```python
y_prediction = Classifier.predict(x_test)
```

```python
#To Determinen accuracy
import sklearn.metrics as metrics 
metrics.accuracy_score(y_test,y_prediction)*100
```

**When we are using  split 50.0% train, remainder test data we get maximum accuracy when we are using K-nn Algorithm**

```python
#confusion matrix
from sklearn.metrics import confusion_matrix
c_matrix = confusion_matrix(y_test,y_prediction)
c_matrix
```

```python
#predicting data
classify = Classifier.predict([[1.0,0.0,0.0,83,86,False]])
```
